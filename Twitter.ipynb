### We see that it identified symbols like the colon as tokens. We need some pre processing to overcome this.

```python
import re
 
emoticons_str = r"""
    (?:
        [:=;] # Eyes
        [oO\-]? # Nose (optional)
        [D\)\]\(\]/\\OpP] # Mouth
    )"""
 
regex_str = [
    emoticons_str,
    r'<[^>]+>', # HTML tags
    r'(?:@[\w_]+)', # @-mentions
    r"(?:\#+[\w_]+[\w\'_\-]*[\w_]+)", # hash-tags
    r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&amp;+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+', # URLs
 
    r'(?:(?:\d+,?)+(?:\.?\d+)?)', # numbers
    r"(?:[a-z][a-z'\-_]+[a-z])", # words with - and '
    r'(?:[\w_]+)', # other words
    r'(?:\S)' # anything else
]
    
tokens_re = re.compile(r'('+'|'.join(regex_str)+')', re.VERBOSE | re.IGNORECASE)
emoticon_re = re.compile(r'^'+emoticons_str+'$', re.VERBOSE | re.IGNORECASE)
 
def tokenize(s):
    return tokens_re.findall(s)
 
def preprocess(s, lowercase=False):
    tokens = tokenize(s)
    if lowercase:
        tokens = [token if emoticon_re.search(token) else token.lower() for token in tokens]
    return tokens
 
tweet = 'RT @akhiludathu: just an example! :D http://example.com #NLP'
print(preprocess(tweet))
```

```python
import operator 
import json
from collections import Counter

from textblob import TextBlob

from nltk.corpus import stopwords #To remove words which aren't of importance
import string
 
punctuation = list(string.punctuation)
stop = stopwords.words('english') + punctuation + ['rt', 'via', 'RT', '\u0bcd']

sentiments_polarity = []
sentiments_subjectivity =[]
sentiments_PETA = []
fname = 'jallikattu.json'
with open(fname, 'r') as f:
    count_all = Counter()
    for line in f:
        tweet = json.loads(line)
        if tweet['lang']=='en':# Let us onoly consider the English words
            terms_stop = [term for term in preprocess(tweet['text']) if term not in stop]
            if 'PETA' in terms_stop: 
                tw1 = TextBlob(tweet['text'])
                sentiments_PETA.append(tw1.sentiment[0])
            count_all.update(terms_stop)
            tw = TextBlob(tweet['text'])
            sentiments_polarity.append(tw.sentiment[0])
            sentiments_polarity.append(tw.sentiment[1])
                
    # Print the most frequent words
    print(count_all.most_common(20))
```

```python
counts_positive =0
counts_negative =0
for i in range(len(sentiments_polarity)):
    if(sentiments_polarity[i]>0):
        counts_positive = counts_positive + 1
    elif(sentiments_polarity[i]<0):
        counts_negative = counts_negative + 1 
```
